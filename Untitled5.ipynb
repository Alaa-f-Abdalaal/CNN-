```python
# =============================
# 1) Load Libraries
# =============================
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix

# =============================
# 2) Load MNIST Data
# =============================
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

# =============================
# 3) Visualize Class Distribution
# =============================
plt.figure(figsize=(8,4))
sns.countplot(x=y_train)
plt.title("Class Distribution in Training Data")
plt.show()

# =============================
# 4) Reshape & Normalize
# =============================
# MNIST images are (28, 28) → CNN needs (28, 28, 1)
X_train = X_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0

# One-hot encoding labels
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# =============================
# 5) Data Augmentation
# =============================
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1
)
datagen.fit(X_train)

# =============================
# 6) Show Example (array & image)
# =============================
index = 0
print("Label:", y_train[index])
print("Array of the image:\n", X_train[index].reshape(28,28))
plt.imshow(X_train[index].reshape(28,28), cmap='gray')
plt.title(f"Label: {y_train[index]}")
plt.show()

# =============================
# 7) Build CNN Model
# =============================
model = Sequential([
    Input(shape=(28,28,1), name="Input_layer"),
    Conv2D(filters=32, kernel_size=3, strides=(1,1), padding='valid', activation='relu', name="conv_layer_1"),
    MaxPooling2D(pool_size=2, name="pooling_1"),

    Conv2D(filters=64, kernel_size=3, activation='relu', name="conv_layer_2"),
    MaxPooling2D(pool_size=2, name="pooling_2"),

    Conv2D(filters=128, kernel_size=3, activation='relu', name="conv_layer_3"),
    Flatten(name="flattening_layer"),

    Dense(units=64, activation='relu'),
    Dense(units=10, activation='softmax', name='output_layer')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# =============================
# 8) Train Model
# =============================
history = model.fit(datagen.flow(X_train, y_train_cat, batch_size=64),
                    epochs=5,
                    validation_data=(X_test, y_test_cat))

# =============================
# 9) Evaluate Model
# =============================
loss, acc = model.evaluate(X_test, y_test_cat)
print(f"Test Accuracy: {acc:.4f}")

# =============================
# 10) Predictions & Confusion Matrix
# =============================
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

print(classification_report(y_test, y_pred_classes))

cm = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# =============================
# 11) Test on a Single Picture
# =============================
img = X_test[0].reshape(1,28,28,1)
pred = model.predict(img)
print("Predicted class:", np.argmax(pred))
plt.imshow(X_test[0].reshape(28,28), cmap='gray')
plt.title(f"Prediction: {np.argmax(pred)}")
plt.show()
```
